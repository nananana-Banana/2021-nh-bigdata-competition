## CusInfo Merge

```python
df = pd.merge(left = train_df, right = cus, how='left', on='act_id')
```

```python
df.columns
```

### 분석 데이터 : 보유기간/주가/nsi/매도날짜/투자성향

```python
clus_data = df[['act_id', 'byn_dt', 'hold_d', 'ivs_icn_cd', 'nsi', 'open_d', 
                'open_mean', 'high_mean', 'low_mean', 'close_mean']]
```

```python
clus_data.info()
```

# <보유기간 - nsi>

```python
df = clus_data[['hold_d', 'nsi']]
```

```python
sns.scatterplot(x="hold_d", y="nsi", data=df, palette="Set2")
```

### K-Means Cluster

```python
ks = range(1, 10)
inertias = []

for k in ks:
    model = KMeans(n_clusters = k)
    model.fit(df)
    inertias.append(model.inertia_)
    
plt.figure(figsize = (4, 4))
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('intertia')
plt.xticks(ks)
plt.show()
```

```python
# k-means clustering 실행
kmeans = KMeans(n_clusters=4)
kmeans.fit(df)

# 결과 확인
result_by_sklearn = df.copy() 
result_by_sklearn["cluster"] = kmeans.labels_
result_by_sklearn.head()
```

```python
sns.scatterplot(x="hold_d", y="nsi", hue="cluster", data=result_by_sklearn, palette="Set2")
```
