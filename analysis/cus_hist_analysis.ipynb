### 종목 코드 / 날짜를 오름차순으로 정렬
- 같은 종목 내에서 날짜가 (예전 -> 최근) 순으로 정렬한다.

```python
sorted_df = df.sort_values(by=['iem_cd', 'bse_dt'], ascending = [True, True])
sorted_df = sorted_df.reset_index(drop = True)
sorted_df['past_d'] = 0
sorted_df
```

### 한 사람의 주식 종목 별 보유 기간을 측정
- 종목 별로 날짜가 정렬되어 있으니깐 위에서부터 스캔해가며 보유기간을 계산한다.
- 계산된 보유기간은 past_d(임시로)에 기록

```python
cal_dur = {}
for i in range(len(sorted_df)):
#     print(i)
    # 위에서부터 하나씩 읽어가는데 딕셔너리에 종목코드가 존재하지 않는다면 해당 날짜 추가
    if sorted_df['iem_cd'][i] not in cal_dur:
        cal_dur[sorted_df['iem_cd'][i]] = sorted_df['bse_dt'][i]
    # 이미 존재하면 해당 잔고가 0인지 확인
    else :
        # 일단 보유기간은 사고/팔고 개념이 있으니깐 팔지 않은 주식은 계산 X
        # 해당 잔고가 0이면 그 날짜와 딕셔너리에 존재하는 날짜를 사용해 보유 기간 계산
        if sorted_df['bnc_qty'][i] == 0:
            sell = sorted_df['bse_dt'][i]
            buy = cal_dur[sorted_df['iem_cd'][i]]
            dur = sell - buy
            sorted_df['past_d'][i] = dur
#             print(sorted_df['iem_cd'][i], ' 의 기간: ', dur)
        # 0이 없다: 다 팔지 않음 ==> 일단 보유기간 측정 안하는 걸로.
        else:
            continue
sorted_df
```

### 모든 계좌번호 종목 별 보유기간 계산


```python
# 잘되나 2개종목만 테스트
# # df = hist[(hist['act_id'] == '25925195f81f0e67e6d983ddd890cd30456f38cb4a3926d4f9d831b57995b021')][['act_id', 'iem_cd', 'bse_dt', 'bnc_qty', 'tot_aet_amt']]
# df2 = hist[(hist['act_id'] == '56dabefb60057f19ee217e3c2aa881627ac213df8fc31d1737af25da9a42a045')][['act_id', 'iem_cd', 'bse_dt', 'bnc_qty', 'tot_aet_amt']]
# # id_cd_list = pd.concat([df, df2], axis = 0)
# # id_cd_list = id_cd_list.reset_index(drop = True)
# test1 = id_cd_list[(id_cd_list['act_id'] == '25925195f81f0e67e6d983ddd890cd30456f38cb4a3926d4f9d831b57995b021')][['act_id', 'iem_cd_list']]
# test2 = id_cd_list[(id_cd_list['act_id'] == '56dabefb60057f19ee217e3c2aa881627ac213df8fc31d1737af25da9a42a045')][['act_id', 'iem_cd_list']]
# id_cd_list = pd.concat([test1, test2], axis = 0)
# id_cd_list = id_cd_list.reset_index(drop = True)
```

```python
def working_day(s_dt, e_dt) : # 공휴일 제외
    s_dt = str(s_dt).split('.')[0]
    e_dt = str(e_dt).split('.')[0]
    s_dt = s_dt[:4] + '-' + s_dt[4:6] + '-' + s_dt[6:]
    e_dt = e_dt[:4] + '-' + e_dt[4:6] + '-' + e_dt[6:]
    return np.busday_count(s_dt, e_dt, weekmask='1111100')
```

```python
def cal_duration(act_id):
    df = hist[(hist['act_id'] == act_id)][['act_id', 'iem_cd', 'bse_dt', 'bnc_qty', 'tot_aet_amt']]
    sorted_df = df.sort_values(by=['iem_cd', 'bse_dt'], ascending = [True, True])
    sorted_df = sorted_df.reset_index(drop = True)
    sorted_df['past_d'] = 0
    
    cur_dur = {}
#     dur_list = []
    for i in range(len(sorted_df)):    
        # 위에서부터 하나씩 읽어가는데 딕셔너리에 종목코드가 존재하지 않는다면 해당 날짜 추가
        if sorted_df['iem_cd'][i] not in cur_dur:
            cur_dur[sorted_df['iem_cd'][i]] = sorted_df['bse_dt'][i]
        # 이미 존재하면 해당 잔고가 0인지 확인
        else :
            # 일단 보유기간은 사고/팔고 개념이 있으니깐 팔지 않은(0이 아닌) 주식은 계산 X
            # 해당 잔고가 0이면 그 날짜와 딕셔너리에 존재하는 날짜를 사용해 보유 기간 계산
            # 어떤 종목을 다 팔고 다시 사고/팔았을 경우도 생각
            if sorted_df['bnc_qty'][i] == 0:
                sell = sorted_df['bse_dt'][i]
                buy = cur_dur[sorted_df['iem_cd'][i]]
                dur = working_day(buy, sell)
                sorted_df['past_d'][i] = dur 
                # 다 팔았으니깐 딕셔너리에서 해당 종목코드 삭제(나중에 다시 살 경우 대비)
                del cur_dur[sorted_df['iem_cd'][i]]
            # 0이 없다: 다 팔지 않음 ==> 일단 보유기간 측정 안하는 걸로.
            else:
                continue
    return sorted_df
```
