# 사는 날 기준으로 volume, open, chnage 추가

```python
import sys, os
sys.path.append(os.pardir)
import numpy as np
import pandas as pd
import fnmatch
import re
```

```python
test_path = 'test_past_d.csv'
```

```python
test_df = pd.read_csv(test_path)
```

```python
test_df.head()
```

```python
cd=test_df['iem_cd']
date=test_df['byn_dt']
```

```python
cd
```

```python
di = './crawling/stockdata'
file_list = os.listdir(di)
```

```python
f_list = []
for i in cd:
    i = i[1:]
    f_name = list(filter(lambda x: i in x, file_list))
    if len(f_name) != 0:
        f_list.append(f_name)
    else:
        f_list.append(None)
```

```python
len(f_list)
```

```python
result = pd.DataFrame(columns=['Open','Volume','Change'])
for i in range(len(f_list)):
    if f_list[i] is not None:
        f = f_list[i][0]
        print(i, f)
        filepath = '{}/{}'.format(di, f)
        df = pd.read_csv(filepath)
        d = str(date[i])
        d = d[:4]+'-'+ d[4:6] +'-'+ d[6:]
        temp = df[df['Date']==d]
        temp = temp.loc[:, ['Open','Volume','Change']]
        result = result.append(temp, ignore_index=True)
    else:
        result = result.append({'Open':0, 'Volume':0, 'Change':0}, ignore_index=True)
```

```python
result.head()
```

```python
result.columns = ['open_d','volume_d','change_d']
```

```python
result
```

```python
test_df = pd.concat([test_df, result], axis=1)
```

```python
test_df
```

```python
output_path = './test_past_d_ovc.csv'
test_df.to_csv(output_path,index = False)
```
