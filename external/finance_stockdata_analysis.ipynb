```python
import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
di = './crawling/stockdata'
file_list = os.listdir(di)
```

```python
file_list
```

```python
code_list = []
for i in file_list:
    code_list.append(i[:6])
```

```python
code_list
```

```python
code_list.index('00088K')
```

```python
len(code_list)
```

## 전체 평균 구하기

```python
# column - Date,Open,High,Low,Close,Volume,Change
result = []
for i in file_list:
    filepath = '{}/{}'.format(di, i)
    df = pd.read_csv(filepath)
    df = df.iloc[:,1:] # 날짜 제외하고 가져오기
    mean_list = list(df.mean())# 평균 구하기
    result. append(mean_list)
```

```python
filepath = '{}/{}'.format(di, i)
df = pd.read_csv(filepath)
```

```python
df
```

```python
len(result)
```

```python
resultdf = pd.DataFrame(result, columns=['open','high','low','close','volume','change'])
resultdf.head(10)
```

```python
resultdf.insert(0,'code',code_list)
resultdf.head(10)
```

```python
output_path = './result/mean_stockdata.csv'
resultdf.to_csv(output_path,index = False)
```

## 전체 평균 구하기 (change 절대값 처리)

```python
# column - Date,Open,High,Low,Close,Volume,Change
result2 = []
change_mean = []
for i in file_list:
    filepath = '{}/{}'.format(di, i)
    df2 = pd.read_csv(filepath)
    change_list = df2.iloc[:, 6].abs() # change 절대값 
    change_mean.append(change_list.mean()) # 절대값들의 평균

    df2 = df2.iloc[:,1:6] # 날짜, change 제외하고 가져오기
    mean_list = np.array(df2.mean())# 평균 구하기
    print(mean_list)
    result2.append(list(mean_list))
```

```python
# column - Date,Open,High,Low,Close,Volume,Change
result2 = []
change_mean = []
for i in file_list:
    filepath = '{}/{}'.format(di, i)
    df2 = pd.read_csv(filepath)
    change_list = df2.iloc[:, 6] # change 절대값 
    change_mean.append(change_list.var()) # 절대값들의 평균

    df2 = df2.iloc[:,1:6] # 날짜, change 제외하고 가져오기
    mean_list = np.array(df2.mean())# 평균 구하기
    print(mean_list)
    result2.append(list(mean_list))
```

```python
len(result2)
```

```python
resultdf2 = pd.DataFrame(result2, columns=['open','high','low','close','volume'])
resultdf2.head(10)
```

```python
resultdf2['abs_change'] = change_mean
resultdf2['sq_abs_change'] = resultdf2['abs_change']**2
resultdf2['sq_abs_change100000'] = resultdf2['sq_abs_change']*100000
```

```python
resultdf2['var_change'] = change_mean
```

```python
resultdf2.insert(0,'code',code_list)
resultdf2.head(10)
```

```python
output_path = './result/var_stockdata.csv'
resultdf2.to_csv(output_path,index = False)
```

## 전체 평균 구하기 (change 제곱 평균 처리)

```python
# column - Date,Open,High,Low,Close,Volume,Change
result3 = []
change_sq_mean = []
for i in file_list:
    filepath = '{}/{}'.format(di, i)
    df3 = pd.read_csv(filepath)
    change_list = df3.iloc[:, 6]**2 # change 절대값 
    change_sq_mean.append(change_list.mean()) # 절대값들의 평균

    df3 = df3.iloc[:,1:6] # 날짜, change 제외하고 가져오기
    mean_list = np.array(df3.mean())# 평균 구하기
    result3.append(list(mean_list))
```

```python
len(change_sq_mean )
```

```python
change_sq_mean 
```

```python
resultdf3 = pd.DataFrame(result3, columns=['open','high','low','close','volume'])
resultdf3.head(10)
```

```python
resultdf3['sq_change'] = change_sq_mean 
resultdf3['sq_change10000'] = resultdf3['sq_change']*100000
```

```python
resultdf3.insert(0,'code',code_list)
resultdf3.head(10)
```

```python
output_path = './result/sq_mean_stockdata.csv'
resultdf3.to_csv(output_path,index = False)
```

# 년도별 평균 구하기
## 2016

```python
# column - Date,Open,High,Low,Close,Volume,Change
result = []
change_mean = []
for i in file_list:
    filepath = '{}/{}'.format(di, i)
    df = pd.read_csv(filepath)
    df = df[df.Date.str.contains(r'2016')] #2016 데이터만 가져오기

    change_list = df.iloc[:, 6].abs() # change 절대값 
    change_mean.append(change_list.mean()) # 절대값들의 평균

    if len(df) != 0:
        df = df.iloc[:,1:6] # 날짜, change 제외하고 가져오기
        mean_list = np.array(df.mean())# 평균 구하기
        result.append(list(mean_list))
    else:
        result.append([0,0,0,0,0])
```

```python
len(result)
```

```python
resultdf = pd.DataFrame(result, columns=['open','high','low','close','volume'])
resultdf.head(10)
```

```python
resultdf['abs_change'] = change_mean
resultdf.head(10)
```

```python
resultdf.insert(0,'code',code_list)
resultdf.head(10)
```

```python
output_path = './result/mean_stock2016.csv'
resultdf.to_csv(output_path,index = False)
```
