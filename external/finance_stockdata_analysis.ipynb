```python
import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dir = './crawling/stock_data'
file_list = os.listdir(dir)
```

```python
file_list
```

```python
# column - Date,Open,High,Low,Close,Volume,Change
result = []
for i in file_list:
    filepath = '{}/{}'.format(dir, i)
    df = pd.read_csv(filepath)
    df = df.iloc[:,1:] # 날짜 제외하고 가져오기
    mean_list = list(df.mean())# 평균 구하기
    result. append(mean_list)
```

```python
result
```

```python
len(result)
```

```python
resultdf = pd.DataFrame(result, columns=['oepn','high','low','close','volume','chage'])
resultdf.head(10)
```

```python
code_path = './crawling/data/code.csv'
code_list = pd.read_csv(code_path)
```

```python
resultdf.insert(0,'code',code_list)
resultdf.head(10)
```

```python
output_path = './result/mean_stockdata.csv'
resultdf.to_csv(output_path,index = False)
```
