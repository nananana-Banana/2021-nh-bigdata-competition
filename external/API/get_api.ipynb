### 종목코드/날짜 별 주당배당금 Load

```python
# train 데이터에 존재하는 종목 이름 load --> 종목 3079개
iem_name_list = list(iem_krl_nm['iem_krl_nm'])
```

```python
api_df = pd.DataFrame({'stckIssuCmpyNm':['Test'],
                       'isinCdNm':['Test'], 
                       'dvdnBasDt':[0], 
                       'cashDvdnPayDt':[0], 
                       'stckGenrDvdnAmt':[0], 
                       'stckGenrCashDvdnRt':[0], 
                       'stckGenrDvdnRt':[0], 
                       'cashGrdnDvdnRt':[0], 
                       'stckGrdnDvdnRt':[0], 
                       'basDt':['Test'] })
api_df
```

```python
# iem_name_list = ['동남합성']
```

```python
for i in notebook.tqdm(range(len(iem_name_list))):
    cd_name = iem_name_list[i]
    f_params ={'serviceKey' :api_key_decode,
         'pageNo' : '', 
         'numOfRows' : '', 
         'resultType' : 'json', 
         'basDt' : '', 
         'crno' : '', 
         'stckIssuCmpyNm' : cd_name }
    response = requests.get(url, params=f_params).json()
    totalCount = response['response']['body']['totalCount']
    
    s_params ={'serviceKey' :api_key_decode,
         'pageNo' : '', 
         'numOfRows' : totalCount, 
         'resultType' : 'json', 
         'basDt' : '', 
         'crno' : '', 
         'stckIssuCmpyNm' : cd_name }
    response = requests.get(url, params=s_params).json()
    totalCount = response['response']['body']['totalCount']
    values = list(response['response']['body']['items'].values())
    all_values = values[0]

    for j in range(len(all_values)):
        stckIssuCmpyNm = all_values[j]['stckIssuCmpyNm'] # 주식발행사의 명칭
        dvdnBasDt = all_values[j]['dvdnBasDt'] # 배당(주식을 가지고 있는 사람들에게 그 소유 지분에 따라 기업이 이윤을 분배하는 것) 기준 일자
        stckGenrDvdnAmt = all_values[j]['stckGenrDvdnAmt'] # 1주당 현금 배당 금액
        stckGenrCashDvdnRt = all_values[j]['stckGenrCashDvdnRt'] # 주식의 현금 배당 비율(권리발행의 현금배정비율)
        new_data = {'stckIssuCmpyNm':stckIssuCmpyNm, 
                       'dvdnBasDt':dvdnBasDt, 
                       'stckGenrDvdnAmt':stckGenrDvdnAmt, 
                       'stckGenrCashDvdnRt':stckGenrCashDvdnRt}
        api_df = api_df.append(new_data, ignore_index = True)
    
```

```python
for i in range(len(api_df)):
    name = api_df['stckIssuCmpyNm'][i]
    if name in iem_dict:
        api_df['iem_cd'][i] = iem_dict[name]
    else:
        api_df['iem_cd'][i] = '0'
```

```python
t_api_df = api_df[api_df.dvdnBasDt >= 20160101] 
```

```python
api_df = t_api_df[t_api_df.dvdnBasDt <= 20210101] 
```

```python
api_df = api_df.reset_index(drop = True)
```

```python
api_df
```

```python
train_df_date = train_df[['iem_cd', 'byn_dt']]
```

```python
train_df_date = train_df_date.sort_values(by=['iem_cd', 'byn_dt'], ascending = [True, True]).reset_index(drop = True)
```

```python
train_df_date
```

```python
train_df_date['stckGenrCashDvdnRt'] = 0
```

```python
train_df_date['stckGenrDvdnAmt'] = 0
```

```python
for i in notebook.tqdm(range(len(train_df_date))):
    code = train_df_date['iem_cd'][i]
    tmp_df = api_df[api_df.iem_cd == code]
    days = sorted(set(tmp_df['dvdnBasDt']))
    for d in range(len(days)):
        if train_df_date['byn_dt'][i] <= int(days[d]):
            index = tmp_df.index[(tmp_df['dvdnBasDt'] == days[d])]
            train_df_date['stckGenrCashDvdnRt'][i] = tmp_df['stckGenrCashDvdnRt'][index]
            train_df_date['stckGenrDvdnAmt'][i] = tmp_df['stckGenrDvdnAmt'][index]
            break
        else:
            continue
```
